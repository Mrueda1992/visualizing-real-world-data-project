```python
from pymongo import MongoClient
import pandas as pd
import re
import requests


client = MongoClient('mongodb://localhost:27017/')
db = client.company
```

```python
#filter companies that have coordinates

query= db.company.find({
    "$and":
    [{"offices.longitude":{"$exists":True}},
     {"offices.latitude":{"$ne":"null"}},
     {"offices.longitude":{"$ne":None}},
     {"offices.longitude":{"$exists":True}
     }
    ]
}
)
```

```python
def dropSpaces(currency):
    return currency.rstrip()
    

with_coords2['total_money_raised'] = with_coords2['total_money_raised'].apply(dropSpaces)
```

```python
def coinsAbsoluteValues(currency):
    currency = str(currency)
    if currency.endswith("M"):
        currency= currency.replace("M","")
        currency= float(currency)*1000000
    if currency.endswith("B"):
        currency= currency.replace("B","")
        currency= float(currency)*1000000000
    if currency.endswith("k"):
        currency= currency.replace("k","")
        currency= float(currency)*1000
    return currency

with_coords2['total_money_raised'] = with_coords2['total_money_raised'].apply(coinsAbsoluteValues)

with_coords2.head()
```
